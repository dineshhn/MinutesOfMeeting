```python
def transcribe_audio(file, model_name):
# ^ Explanation:
# Defines a function named transcribe_audio that takes in two parameters:
# file: The audio file to transcribe (either file path or raw audio).
# model_name: The transcription model key (e.g., "whisper-tiny.en") to select from the predefined model dictionary.

    pipe = transcription_pipelines.get(model_name)

    # ^ Explanation:
    # Retrieves the corresponding model pipeline from the transcription_pipelines dictionary using the model_name.
    # This pipe will be a Hugging Face pipeline for automatic speech recognition (ASR).

    if pipe is None:
        raise ValueError("Unsupported model")

    # ^ Explanation:
    # Checks if the given model name was not found in the dictionary.
    # If it’s missing, it raises a ValueError to alert that an unsupported model was requested.

    start = time.time()

    # ^ Explanation:
    # Records the current timestamp before transcription starts.
    # This is used later to calculate the time taken for transcription.

    if "whisper" in model_name:
        result = pipe(file, return_timestamps=True)

    # ^ Explanation:
    # Special handling for Whisper models.
    # If the model name contains "whisper", it invokes the pipeline with return_timestamps=True to also fetch word-level timestamps.
    # Whisper models support this argument natively.

    else:
        result = pipe(file)

    # ^ Explanation:
    # For non-Whisper models (e.g., Wav2Vec2), just call the pipeline normally without return_timestamps.

    end = time.time()

    # ^ Explanation:
    # Records the timestamp after the transcription finishes to compute the total processing time.

    return result["text"], f"Transcription time: {end - start:.2f} seconds"

    # ^ Explanation:
    # Returns a tuple:
    # The transcribed text from the audio.
    # A string that displays the transcription duration, rounded to two decimal places.


# Summary:
# This function selects the appropriate ASR model, runs transcription on the input file (with Whisper-specific options if needed), and returns both the transcribed text and the time it took to process.
```

```python
import torch
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import textwrap
import time
import gc

# ^ Explanation:
# Imports necessary modules:
# torch: for handling PyTorch tensors and device allocation (CPU/GPU).
# transformers: Hugging Face library for model pipelines and loading models/tokenizers.
# textwrap: for splitting large text into smaller chunks.
# time: to measure the duration of summarization.
# gc: for garbage collection, helps free up memory.

def summarize_text(text, model_name):

# ^ Explanation:
# Defines the summarize_text function that:
# Accepts text to summarize.
# Uses the specified model_name (e.g., a BART or Pegasus summarization model).

    gc.collect()

    # ^ Explanation:
    # Manually runs garbage collection to clean up unused memory before starting the task.

    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    # ^ Explanation:
    # If a CUDA-capable GPU is available, clears the GPU memory cache to prevent memory overload.

    device = 0 if torch.cuda.is_available() else -1
    device_str = "cuda" if device == 0 else "cpu"
    print(f"Device set to use: {device_str}")

    # ^ Explanation:
    # Sets the device:
    # 0 if CUDA (GPU) is available.
    # -1 for CPU.
    # Also prints which device is being used for debugging.

    try:

        tokenizer = AutoTokenizer.from_pretrained(model_name)
        model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to("cpu")

    # ^ Explanation:
    # Loads the tokenizer and model from Hugging Face for the given model name.
    # Explicitly pushes the model to CPU using .to("cpu").


        summarizer = pipeline(
            "summarization",
            model=model,
            tokenizer=tokenizer,
            device=-1
        )

        # ^ Explanation:
        # Initializes a Hugging Face pipeline for summarization using the previously loaded model and tokenizer.
        # device=-1 forces it to run on CPU regardless of earlier settings.

    except Exception as e:
        print(f"Error loading model {model_name}: {e}")
        return "[Error loading model]", 0.0

    # ^ Explanation:
    # Catches and logs model loading errors and returns a fallback summary and 0 seconds duration.


    max_chunk_size = 900
    chunks = textwrap.wrap(text, max_chunk_size)

    # ^ Explanation:
    # Sets maximum chunk size to 900 characters.
    # Splits the input text into chunks of ~900 characters to handle long text (most summarizers have a limit on input length).

    final_summary = ""
    start_time = time.time()

    # ^ Explanation:
    # Initializes an empty string for the final summary and records the start time of the summarization process.

    for i, chunk in enumerate(chunks): # Explanation: Loops through each chunk of text for summarization.
        print(f"Summarizing chunk {i+1}/{len(chunks)} with {model_name}") # Explanation: Logs progress of which chunk is being summarized.
        try:
            if chunk.strip() and len(chunk.strip().split()) > 5: # Explanation: Skips chunks that are empty or too short (5 or fewer words), which are unlikely to need summarization.
                output = summarizer(
                    chunk,
                    max_length=130,
                    min_length=30,
                    do_sample=False
                )

            # ^ Explanation:
            # Calls the summarization pipeline:
            # max_length=130, min_length=30: Controls output size.
            # do_sample=False: Makes generation deterministic (not random).

                summary_text = output[0]['summary_text'].strip() # Explanation: Extracts and strips the actual summary text from the result.

                if "CNN.com will feature" not in summary_text:
                    final_summary += summary_text + " "
                else:
                    print("Detected fallback summary, skipping.")

                # ^ Explanation:
                # Skips generic, fallback-style summaries often returned when the model fails to produce meaningful output (like CNN.com templates).

            else:
                print("Skipping empty or short chunk.") # Explanation: Handles skipped chunks with logging for visibility.
        except Exception as e:
            print(f"Error summarizing chunk with {model_name}: {e}")
            final_summary += "[Error summarizing chunk] "

        # ^ Explanation:
        # Logs any chunk-level summarization failure and appends an error placeholder to the final summary.

    end_time = time.time()
    elapsed_time = round(end_time - start_time, 2) # Explanation: Measures and rounds off total summarization time in seconds.

    return final_summary.strip(), elapsed_time # Explanation: Returns the complete summarized text and total time taken to generate it.

# Summary:
# This function handles robust, chunked text summarization using Hugging Face models.
# It includes device setup, error handling, chunk skipping, and result sanitization — making it production-ready and memory-efficient.
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# ^ Explanation:
# Imports common classification evaluation metrics from scikit-learn:
# precision, recall, f1, and accuracy will be used on binary vector comparisons of words.

from rouge_score import rouge_scorer # Explanation: Imports the ROUGE metric scorer from rouge_score, used for text similarity evaluation.

import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

# ^ Explanation:
# Imports NLTK (Natural Language Toolkit) for:
# BLEU score computation (sentence_bleu)
# Smoothing to handle short sequences and avoid zero scores.

from sklearn.feature_extraction.text import CountVectorizer

# ^ Explanation:
# Used to vectorize the reference and predicted texts into binary bag-of-words vectors for precision/recall/f1/accuracy metrics.

def evaluate_summary(reference_text, summary):

# ^ Explanation:
# Defines the function that evaluates the similarity and quality of a generated summary against a reference text using multiple metrics.

    nltk.download('punkt', quiet=True)

    # ^ Explanation:
    # Downloads the Punkt tokenizer model from NLTK (if not already downloaded), used for sentence and word tokenization.
    # quiet=True suppresses download logs.


    ref_tokens = nltk.word_tokenize(reference_text.lower())
    pred_tokens = nltk.word_tokenize(summary.lower())

    # ^ Explanation:
    # Tokenizes both reference and predicted texts into lowercase words for uniform comparison.
    # Required for BLEU and ROUGE.


    bleu = sentence_bleu([ref_tokens], pred_tokens, smoothing_function=SmoothingFunction().method1)

    # ^ Explanation:
    # Calculates the BLEU score between the reference and predicted token lists.
    # Uses smoothing to avoid zero scores when n-gram matches are sparse (e.g., for short summaries).


    scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
    rouge_scores = scorer.score(reference_text, summary)

    # ^ Explanation:
    # Creates a ROUGE scorer instance to calculate ROUGE-1 and ROUGE-L metrics.
    # use_stemmer=True makes it more robust by reducing words to their stems (e.g., "running" → "run").
    # Applies scoring to the full reference and summary texts.


    vectorizer = CountVectorizer(binary=True)
    vectors = vectorizer.fit_transform([reference_text, summary]).toarray()
    y_true, y_pred = vectors[0], vectors[1]

    # ^ Explanation:
    # Converts both texts into binary feature vectors (presence/absence of words).
    # vectors[0] is for reference (true values), vectors[1] is for summary (predicted values).
    # This lets us use classification metrics like precision, recall, etc., by treating word presence as binary classification.


    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    accuracy = accuracy_score(y_true, y_pred)

    # ^ Explanation:
    # Computes standard classification metrics on the binary vectors:
    # zero_division=0 avoids division-by-zero errors and returns 0 when no positives are predicted/found.

    return {
        "BLEU": round(bleu, 2),
        "ROUGE1": round(rouge_scores["rouge1"].fmeasure, 2),
        "ROUGEL": round(rouge_scores["rougeL"].fmeasure, 2),
        "PRECISION": round(precision, 2),
        "RECALL": round(recall, 2),
        "F1": round(f1, 2),
        "ACCURACY": round(accuracy, 2),
    }

    # ^ Explanation:
    # Returns a dictionary of all computed metrics, rounded to two decimal places for readability.
    # ROUGE1 and ROUGEL return the F1 variant of the ROUGE score.

# Summary:
# This function provides a comprehensive evaluation of how similar the summary is to the reference text using:
# BLEU (n-gram overlap),
# ROUGE (recall-oriented overlap),
# and classification-style metrics (precision, recall, F1, accuracy) using bag-of-words vectorization.
```

```python
def format_action_items(action_items_by_model):

# ^ Explanation:
# Defines a function that takes a dictionary (action_items_by_model) where:
# Keys are model names (e.g., "facebook/bart-large-cnn")
# Values are lists of extracted action items for each model

    formatted = "<h4>Action Items</h4>" # Explanation: Initializes an HTML string with a heading for the action items section.
    for model, items in action_items_by_model.items(): # Explanation: Iterates through each model and its corresponding action item list.
        formatted += f"<b>{model}:</b><br>" # Explanation: Adds the model name in bold followed by a line break in the HTML output.
        if items: # Explanation: Checks if the list of action items for this model is non-empty.
            formatted += "<ul>" + "".join(f"<li>{item}</li>" for item in items) + "</ul>" # Explanation: If action items exist, wrap them in an unordered HTML list (<ul>) with each item as a list element (<li>).
        else:
            formatted += "<i>No action items found.</i><br>" # Explanation: If there are no action items for the model, show a message in italic stating that no items were found.
    return formatted # Explanation: Returns the complete HTML-formatted string containing all the action items grouped by model.
```

```python
def extract_action_items(summary, keywords=None):

# ^ Explanation:
# Defines a function that attempts to extract action items from a given summary using a list of specified keywords.
# keywords is an optional parameter; if not provided, default keywords will be used.

    if not keywords:
        keywords = ["must", "should", "need to", "expanding", "revenue", "plan", "action"]

    # ^ Explanation:
    # If the user doesn't provide a keywords list, a default set of action-indicating words/phrases is used.

    keywords = [kw.lower() for kw in keywords] # Explanation: Converts all keywords to lowercase for case-insensitive matching.
    sentences = summary.split(". ")

    # ^ Explanation: Splits the summary into sentences using . (period + space) as a delimiter.
    # This is a simple way to separate sentences, assuming each sentence ends with a period followed by a space.

    return [s for s in sentences if any(kw in s.lower() for kw in keywords)]

    # ^ Explanation:
    # Returns a list of sentences from the summary that contain at least one of the action-related keywords (case-insensitive).
    # Uses list comprehension with a condition that checks if any keyword appears in each sentence.
```

```python
from fpdf import FPDF
import os
import datetime

# ^ Explanation:
# Imports the necessary modules.
# FPDF is used to generate PDF documents.
# os is used to work with file paths.
# datetime is used to generate a timestamp for the filename.

def generate_pdf(transcription, results, graph_paths):

# ^ Explanation:
# Defines the function generate_pdf which takes in:
# transcription: the full transcribed text.
# results: a dictionary containing summaries, action items, and evaluation metrics for each summarizer model.
# graph_paths: file paths to saved graph images (e.g., bar, pie, line, radar charts).


    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", 'B', 16)
    pdf.cell(0, 10, "Minutes of Meeting Report", ln=True)

    # ^ Explanation:
    # Creates a new PDF document, adds a page, sets the font, and adds a bold title for the report.


    pdf.set_font("Arial", 'B', 14)
    pdf.cell(0, 10, "Transcribed Text", ln=True)
    pdf.set_font("Arial", size=11)
    pdf.multi_cell(0, 10, transcription)

    # ^ Explanation:
    # Adds a section title for the transcribed text.
    # Sets a smaller font and prints the transcribed text using multi_cell for line wrapping.


    for model, data in results.items(): # Explanation: Loops over each model in the results dictionary to generate summaries, action items, and evaluation metrics.
        pdf.set_font("Arial", 'B', 14)
        pdf.cell(0, 10, f"Summary - {model}", ln=True)
        pdf.set_font("Arial", size=11)
        pdf.multi_cell(0, 10, data['summary'])

        # ^ Explanation:
        # Adds the summary title for the current model.
        # Writes the actual summary in a wrapped format.

        pdf.set_font("Arial", 'B', 12)
        pdf.cell(0, 10, "Action Items", ln=True)
        pdf.set_font("Arial", size=11)
        for item in data['actions']:
            pdf.cell(5)
            pdf.multi_cell(0, 10, f"- {item}")

        # ^ Explanation:
        # Adds a title and then lists each action item as a bullet point with indentation.

        pdf.set_font("Arial", 'B', 12)
        pdf.cell(0, 10, "Evaluation Metrics", ln=True)
        pdf.set_font("Arial", size=11)
        for k, v in data["metrics"].items():
            pdf.cell(0, 10, f"{k.upper()}: {v:.2f}", ln=True)

        # ^ Explanation:
        # Displays evaluation metrics such as BLEU, ROUGE, etc., with two decimal precision.


    titles = [
        "Bar Chart - Evaluation Metrics",
        f"Pie Chart - {list(results.keys())[0]}",
        f"Pie Chart - {list(results.keys())[1]}" if len(results) > 1 else "Pie Chart",
        "Line Chart - Evaluation Metrics",
        "Radar Chart - Evaluation Metrics"
    ]

    # ^ Explanation:
    # Creates a list of titles for the 5 charts. The pie chart titles are model-specific.

    for idx, path in enumerate(graph_paths):
        if idx % 2 == 0:
            pdf.add_page()

        pdf.set_font("Arial", 'B', 12)
        pdf.cell(0, 10, titles[idx], ln=True)
        pdf.image(path, w=180)
        pdf.ln(5)

    # ^ Explanation:
    # Loops through each chart image and adds them to the PDF.
    # A new page is added before every two charts.
    # Adds a title above each chart image.

    filename = f"MoM_Report_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.pdf"

    # ^ Explanation:
    # Generates a timestamped filename like MoM_Report_20250420_214512.pdf.

    file_path = os.path.join(os.getcwd(), filename)
    pdf.output(file_path)
    return file_path

    # ^ Explanation:
    # Saves the generated PDF to the current working directory.
    # Returns the full path of the generated PDF.
```

```python
def show_graphs1(results): # Defines the function show_graphs1 which takes results — a dictionary containing evaluation metric scores for each summarizer model — and returns 5 Plotly graph figures.
    import plotly.graph_objects as go # Imports Plotly’s graph_objects module, which allows you to construct complex plots like bar, line, pie, and radar charts.

    models = list(results.keys())
    metrics = list(results[models[0]]["metrics"].keys())

    # ^ Explanation
    # Extracts the list of model names (e.g., BART, Pegasus) from results.
    # Extracts the list of metric names (e.g., BLEU, ROUGE) from the first model’s metric dictionary.

    fig_bar = go.Figure()
    fig_line = go.Figure()

    # ^ Explanation:
    # Initializes empty figures for a bar chart and a line chart.

    for model in models:
        values = list(results[model]["metrics"].values())
        fig_bar.add_trace(go.Bar(name=model, x=metrics, y=values))
        fig_line.add_trace(go.Scatter(name=model, x=metrics, y=values, mode='lines+markers'))

    # ^ Explanation:
    # For each model:
    # Extracts its metric values.
    # Adds a bar trace to fig_bar to show each metric.
    # Adds a line + marker trace to fig_line to visualize metrics over a line graph.

    fig_bar.update_layout(title="Evaluation Metrics - Bar Chart", barmode='group')
    fig_line.update_layout(title="Evaluation Metrics - Line Chart")

    # ^ Explanation:
    # Updates layout for the bar chart to show grouped bars by metric.
    # Sets the title for the line chart.


    fig_radar = go.Figure() # Initializes a radar (polar) chart figure.
    for model in models:
        values = list(results[model]["metrics"].values())
        values += values[:1]
        fig_radar.add_trace(go.Scatterpolar(r=values, theta=metrics + [metrics[0]], fill='toself', name=model))

        # ^ Explanation:
        # For each model:
        # Retrieves metric values and repeats the first value at the end to close the radar shape.
        # Adds a Scatterpolar trace to the radar chart with those values and metric names as angular coordinates.

    fig_radar.update_layout(polar=dict(radialaxis=dict(visible=True, range=[0, 1])), title="Radar Chart") # Sets radar chart layout: visible radial axis with fixed range [0, 1], and adds a title.


    fig_pie1 = go.Figure(go.Pie(
        labels=metrics,
        values=list(results[models[0]]["metrics"].values()),
        title=f"Pie Chart - {models[0]}"
    ))

    # ^ Explanation:
    # Creates a pie chart (fig_pie1) for the first model showing metric contributions.

    fig_pie2 = go.Figure(go.Pie(
        labels=metrics,
        values=list(results[models[1]]["metrics"].values()),
        title=f"Pie Chart - {models[1]}"
    ))

    # ^ Explanation:
    # Creates another pie chart (fig_pie2) for the second model.

    return fig_bar, fig_pie1, fig_pie2, fig_line, fig_radar

    # ^ Explanation:
    # Returns all 5 chart figures to be rendered or exported elsewhere (like in PDF or Gradio interface).
```

```python
def process_and_format_outputs(file, trans_model, keywords, summarizer_models):

# ^ Explanation:
# Defines the main function to handle:
# audio transcription,
# text summarization using selected models,
# evaluation of summaries,
# graph generation,
# PDF report export.
# It supports Gradio generator syntax using yield.

    try:
        print("=== PROCESS STARTED ===")
        print(f"Selected Transcription Model: {trans_model}")
        print(f"Selected Summarizer Models: {summarizer_models}")

    # ^ Explanation:
    # Prints the selected models and starts the log to help debug and trace.


        transcription, trans_time = transcribe_audio(file, trans_model)
        print("Transcribed")

        # ^ Explanation:
        # Transcribes the uploaded/recorded audio using the selected model and captures the transcription time.

        yield transcription, "Generating summary...", "Generating summary...", f"{trans_time}", "", "", {}, None, "Transcription complete."

        # ^ Explanation:
        # Immediately returns interim results to the UI:
        # Shows transcription text.
        # Shows “Generating summary…” in summary boxes.
        # Shows transcription time.
        # Other outputs are temporarily blank or placeholders.


        results = {}
        from concurrent.futures import ThreadPoolExecutor

        # ^ Explanation:
        # Prepares a dictionary to store results and imports thread pool to parallelize summary generation.

        with ThreadPoolExecutor() as executor:
            futures = {
                model: executor.submit(summarize_text, transcription, model)
                for model in summarizer_models
            }

        # ^ Explanation:
        # Launches parallel threads to summarize transcription using each selected model.

            for model, future in futures.items():
                print(f"Summarizing with: {model}")
                summary, summ_time_raw = future.result()

            # ^ Explanation:
            # Waits for each thread to complete and gets the summary and its processing time.


                try:
                    summ_time = float(summ_time_raw)
                except:
                    summ_time = 0.0

                # ^ Explanation:
                # Converts the time string to float safely; if error, sets time to 0.

                print(f"Summary complete for: {model}")
                actions = extract_action_items(summary, keywords)
                metrics = evaluate_summary(transcription, summary)

                # ^ Explanation:
                # Extracts action items based on keywords.
                # Evaluates the summary against transcription using BLEU, ROUGE, etc.

                results[model] = {
                    "summary": summary,
                    "actions": actions,
                    "metrics": metrics,
                    "times": {
                        "transcription": trans_time,
                        "summarization": summ_time
                    }
                }

                # ^ Explanation:
                # Stores all relevant data (summary, action items, metrics, times) in a structured format under results.


        summary1 = results.get(summarizer_models[0], {}).get("summary", "")
        summary2 = results.get(summarizer_models[1], {}).get("summary", "") if len(summarizer_models) > 1 else ""

        # ^ Explanation:
        # Gets summaries from the first and second selected models.

        formatted_actions = format_action_items({m: r["actions"] for m, r in results.items()})

        # ^ Explanation:
        # Prepares HTML-formatted action item text for display.


        metric_keys = list(next(iter(results.values()))["metrics"].keys()) # Extracts metric names from any one model’s result.
        metrics_html = "<h4>Evaluation Metrics</h4><table border='1' cellpadding='6' style='border-collapse: collapse;'>"
        metrics_html += "<tr><th>Metric</th>" + "".join(f"<th>{model}</th>" for model in results.keys()) + "</tr>"

        # ^ Explanation:
        # Starts creating a metrics comparison HTML table.

        for metric in metric_keys:
            metrics_html += "<tr>"
            metrics_html += f"<td><b>{metric}</b></td>"
            for model in results.keys():
                val = results[model]["metrics"].get(metric, 0.0)
                metrics_html += f"<td>{val:.2f}</td>"
            metrics_html += "</tr>"
        metrics_html += "</table>"

        # ^ Explanation:
        # Fills the HTML table row by row, displaying each metric value for every selected model.


        fig_bar, fig_pie1, fig_pie2, fig_line, fig_radar = show_graphs1(results)

        # ^ Explanation:
        # Calls show_graphs1() to generate interactive charts for the evaluation.

        graph_paths = []
        for i, fig in enumerate([fig_bar, fig_pie1, fig_pie2, fig_line, fig_radar], start=1):
            path = f"graph_{i}.png"
            fig.write_image(path)
            graph_paths.append(path)

        # ^ Explanation:
        # Saves all graphs as PNG images for including in the PDF report.


        summarization_times = "\n".join([
            f"Summarization time ({model}): {results[model]['times']['summarization']:.2f} seconds"
            for model in results
        ])
        time_output = f"{trans_time}\n{summarization_times}"

        # ^ Explanation:
        # Formats time string for both transcription and all summarization models for display in the Gradio output box.


        pdf_filename = generate_pdf(
            transcription=transcription,
            results=results,
            graph_paths=graph_paths
        )

        # ^ Explanation:
        # Generates a PDF file combining transcription, summaries, action items, metrics, and the graphs.

        yield transcription, summary1, summary2, time_output, formatted_actions, metrics_html, results, pdf_filename, "Summarization complete."

        # ^ Explanation:
        # Returns all the final values to their corresponding Gradio output boxes.

    except Exception as e:
        error_msg = f"Error during processing: {str(e)}"
        print("EXCEPTION:", error_msg)
        yield error_msg, "", "", "", "", "", {}, None, "Error"

    # ^ Explanation:
    # Catches and reports any exceptions that occurred during processing and safely returns placeholders to Gradio.
```
